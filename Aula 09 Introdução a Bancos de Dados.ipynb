## Usando SQLite

```python
import sqlite3

conn = sqlite3.connect('shoefly.db')

import pandas as pd

df = pd.read_csv('shoefly_orders.csv')
df.set_index('id')

df.to_sql('orders', conn, if_exists='replace', index=False)

conn.close()
```

```python
# tabelas_bd(): retorna um dataframe com as tabelas do banco de dados.

def tabelas_bd():
    query = "SELECT name FROM sqlite_master WHERE type='table'"

    conn = sqlite3.connect('coderhouse.db')
    
    df = pd.read_sql_query(query, conn)

    conn.close()

    return df
```

```python
# salva_bd(df, nome_tabela): salva dataframe df na tabela nome_tabela.

def salva_bd(df, nome_tabela):
    conn = sqlite3.connect('coderhouse.db')

    df.to_sql(nome_tabela, conn, if_exists='replace', index=False)

    conn.close()
```

```python
# carrega_bd(nome_tabela): carrega tabela nome_tabela num dataframe. 

def carrega_bd(nome_tabela):
    conn = sqlite3.connect('coderhouse.db')

    df = pd.read_sql(f'SELECT * FROM {nome_tabela}', conn)

    conn.close()

    return df
```
