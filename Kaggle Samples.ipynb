```python
import pandas as pd

df = pd.read_csv('realtor-data.zip.csv')
df
```

```python
# somente casas for_sale
df = df[df['status'] == 'for_sale']
df
```

```python
# sem duplicidades de status, city, state e price
# priorizando as casas maiores
df = df.sort_values('house_size')
df = df.drop_duplicates(['status', 'city', 'state', 'price'], keep='last')
df
```

```python
# amostra com 1000 registros, sendo 
# 250 do estado de Massachusetts
massachusetts = df[df['state'] == 'Massachusetts']
sample_1 = massachusetts.sample(n=250)
# 250 do estado de Puerto Rico
puerto_rico = df[df['state'] == 'Puerto Rico']
sample_2 = puerto_rico.sample(n=250)
# 250 do estado de Connecticut 
connecticut = df[df['state'] == 'Connecticut']
sample_3 = connecticut.sample(n=250)
# 250 do estado de Rhode Island e New York
new_york = df[(df['state'] == 'Rhode Island') | (df['state'] == 'New York')]
sample_4 = new_york.sample(n=250)

sample = pd.concat([sample_1, sample_2, sample_3, sample_4])
sample
```
